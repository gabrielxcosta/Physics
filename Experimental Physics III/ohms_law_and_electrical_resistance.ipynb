# Experimento I - Lei de Ohm e Resistência Elétrica

### Objetivos

Este experimento tem por objetivo estudar a dependência da diferença de potencial ($V$) com a corrente ($I$) de um dado circuito para componentes ôhmicos e não ôhmicos.

### Introdução

Se um elemento do circuito é percorrido por uma corrente elétrica, estabelece-se uma diferença de potencial $V$ entre seus terminais. Um fato experimental importante que, em boa parte dos casos, a tensão $V$ será proporcional a corrente $I$. Este comportamento pode ser escrito pela equação:

$$V = RI$$

onde $R$ é uma constante de proporcionalidade denominada *resistenca elétrica do fio*, sendo sua unidade no Sistema Internacional (SI) o *ohm* ($\Omega$). Esta equação é mais conhecida como *Lei de Ohm*. A maioria dos metais obedece a esta lei, e, por este motivo, tais materiais são chamados de materials ôhmicos.

Observa-se que a resistência $R$ de um fio condutor é proporcional ao seu comprimento $L$ e inversamente proporcional a sua área $A$, portanto:

$$R = \rho \frac{L}{A},$$

onde a constante de proporcionalidade $\rho$ é chamada de *resistividade* do material e suas unidades no SI são ($\Omega.m$) Os metais e ligas metálicas são os materials com menor resistividade e, por este motivo, os melhores condutores.

Outros materiais apresentam um comportamento substancialmente diferente da Lei de Ohm, neles a dependência entre $V$ e $I$ não é uma relação linear e, razão pela qual são denominados de materiais não-ôhmicos ou não-lineares.

# Importando as bibliotecas necessárias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import warnings as w
```

```python
w.filterwarnings('ignore')
sns.set_palette('Accent')
sns.set_style('darkgrid')
```

# Instanciando os dados

```python
V = np.array([8.2, 21.9, 30.5, 43.3, 61.3, 73.8, 88.7, 98.2, 119.6, 129.0]) # DDP [mV]
I = np.array([1.5, 4.0, 5.7, 8.1, 11.5, 13.8, 16.6, 18.4, 22.4, 24.5]) # Corrente [mA]

len(V) == len(I)
```

```python
data = pd.DataFrame(
    {
        'V' : V,
        'I' : I
    }
)
data.index = range(1, len(data) + 1)
```

**Podemos observar que existe uma relação linear clara entre $I$ e $V$.**

# Rodando a regressão

## Criando uma Series (pandas) para armazenar a DDP (y)

```python
y = data['V'].to_numpy()
y
```

##  Criando uma Series (pandas) para armazenar as variáveis explicativas (X)

```python
X = data['I'].to_numpy().reshape(-1, 1)
X
```

# Criando os datasets de treino e de teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Verificando os tamanhos dos arquivos gerados pela função train_test_split

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.shape[0] + X_test.shape[0]
```

## Instanciando a classe LinearRegression()

```python
model = LinearRegression()
```

## Utilizando o método fit() do objeto "modelo" para estimar nosso modelo linear utilizando os dados de TREINO (y_train e X_train)

```python
model.fit(X_train, y_train)
```

## Obtendo o coeficiente de determinação ($R²$) do modelo estimado com os dados de TREINO

```python
print(f'R² = {model.score(X_train, y_train)}')
```

##  Gerando previsões para os dados de TESTE (X_test) utilizando o método predict() do objeto "modelo"

```python
y_expected = model.predict(X_test) 
```

## Obtendo o coeficiente de determinação ($R²$) para as previsões do nosso modelo

```python
print(f'R² = {metrics.r2_score(y_test, y_expected)}')
```

## Intercepto e o Coeficinte de Regressão do nosso modelo

```python
model.intercept_
```
