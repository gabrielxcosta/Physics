## 1) Introdução

Um sistema quântico de $N$ partículas idênticas pode ser representado pela função de onda

<center>

$\Psi = \Psi(q_{1}, ..., q_{N}),$          **(1)**

</center>

onde $q_{j}$ designa todas as coordenadas da partícula $j$ (posição e spin, por exemplo). No entanto, nem todas as funções de onda desse tipo, que satisfazem a equação de Schroedinger independente do tempo, são representações aceitáveis de um sistema quântico. Exige-se também a propriedade de simetria,

<center>

$\Psi(q_{1}, ..., q_{i}, ..., q_{j}, ..., q_{N}) = \pm \Psi(q_{1}, ..., q_{i}, ..., q_{j}, ..., q_{N}),$          **(2)**

</center>

indicando que o estado quântico do sistema não se altera quando se trocam as coordenadas de duas partículas. As funções de onda simétricas estão associadas a partículas de spin inteiro (fótons, fônons, mágnons, átomos de $He^{4}$). Essas partículas são denominadas **bósons** e obedecem a chamada **estatística de Bose-Einstein**. As funções de onda anti-simétricas estão associadas a partículas de spin semi-inteiro (elétrons, prótons, neutrons, átomos de $He^{3}$). Essas partículas são denominadas **férmions** e obedecem a **estatística de Fermi-Dirac**.

Vamos, por exemplo, considerar um sistema de apenas duas partículas idênticas e independentes (ou seja, praticamente não-interagentes), dado pelo hamiltoniano

<center>

$\cal{H} = \cal{H}_{1} + \cal{H}_{2},$          **(3)**

</center>

onde

<center>

${\cal{H}_{j}} = \frac{1}{2m} \overrightarrow{p_{j}}^{2} + V (\overrightarrow{r_{j}}),$          **(4)**

</center>

para $j = 1$ ou $j = 2$. As autofunções do hamiltoniano $\cal{H}$, correspondentes a uma dada energia $E$, podem ser escritas na forma de um produto, $\psi_{n_{1}} (\overrightarrow{r_{1}}) \psi_{n_{2}} (\overrightarrow{r_{2}})$, tal que


<center>

${\cal{H}_{1}} \psi_{n_{1}} (\overrightarrow{r_{1}}) = \varepsilon_{n_{1}} \psi_{n_{1}} (\overrightarrow{r_{1}})$ **(5)**

</center>

e

<center>

${\cal{H}_{2}} \psi_{n_{2}} (\overrightarrow{r_{2}}) = \varepsilon_{n_{2}} \psi_{n_{2}} (\overrightarrow{r_{2}})$ **(6)**

</center>

com

<center>

$E = \varepsilon_{n_{1}} + \varepsilon_{n_{2}}.$ **(7)**

</center>

Os estados quânticos aceitáveis do sistema, no entanto, são representados pelas combinações lineares simétrica e anti-simétrica desse produto,

<center>

$\Psi_{S}(\overrightarrow{r_{1}}, \overrightarrow{r_{2}}) = \frac{1}{\sqrt{2}} [\psi_{n_{1}}(\overrightarrow{r_{1}}) \psi_{n_{2}}(\overrightarrow{r_{2}}) + \psi{n_{1}}(\overrightarrow{r_{2}}) \psi{n_{2}}(\overrightarrow{r_{1}})]$ **(8)**

</center>

e

<center>

$\Psi_{A}(\overrightarrow{r_{1}}, \overrightarrow{r_{2}}) = \frac{1}{\sqrt{2}} [\psi_{n_{1}}(\overrightarrow{r_{1}}) \psi_{n_{2}}(\overrightarrow{r_{2}}) - \psi{n_{1}}(\overrightarrow{r_{2}}) \psi{n_{2}}(\overrightarrow{r_{1}})].$ **(9)**

</center>

Vamos chamar $\psi_{n} (\overrightarrow{r})$ de **orbital ou estado de partícula única**. Note que $\Psi_{A} = 0$ quando $n_{1} = n_{2}$. Portanto, no caso de férmions não pode haver duas partículas no mesmo orbital (ou seja, com o mesmo conjunto de números quânticos), de acordo com o **princípio de exclusão de Pauli**. Todas as evidências experimentais apoiam essa grande classificação das partículas em bósons (funções de onda simétricas) ou férmions (funções anti-simétricas). Tem havido esforços para a consideração de estatísticas intermediárias, que permanecem, no entanto, como simples especulações matemáticas.

A título de ilustração, vamos agora construir explicitamente os estados quânticos de um sistema de duas partículas idênticas e independentes supondo que os números quânticos dos orbitais ($n_{1}$ e $n_{2}$) possam assumir apenas três valores distintos (que vamos chamar 1, 2 e 3). Temos então as seguintes possibilidades:

(ii) se as partículas forem férmions, vamos ter apenas três estados quânticos acessíveis ao sistema, devidamente indicados na tabela abaixo. </em>

## 2) Orbitais de uma partícula livre

Vamos, inicialmente, considerar uma partícula de massa $m$, em uma dimensão, numa região de comprimento $L$ (isto é, com a coordenada $x$ entre $0$ e $L$). Levando em conta apenas as coordenadas de posição, o orbital $\psi_{n}(x)$ é dado pela solução da equação de Schroedinger,

<center>

${\cal{H}} \psi_{n}(x) = \varepsilon_{n} \psi_{n} (x),$ **(10)**

</center>

onde

<center>

${\cal{H}} = \frac{1}{2m} p^{2} = -\frac{\hbar^{2}}{2m} \frac{d^{2}}{dx^{2}}.$ **(11)**

</center>

Temos, portanto, soluções da forma

<center>

$\psi_{n}(x) = Ce^{ikx}$ com $\varepsilon_{n} = \frac{\hbar^{2} k^{2}}{2m}.$ **(12)**

</center>

A quantização da energia é introduzida por meio das condições de contorno. Por pura conveniência, vamos adotar condições periódicas de contorno (como estamos interessados na conexão com a termodinâmica, que se verifica apenas no limite termodinâmico, as condições de contorno não devem ter qualquer influência sobre os resultados físicos finais). Então, vamos supor que

<center>

$\psi_{n} (x) = \psi_{n} (x + L).$ **(13)**

</center>

Portanto, $exp(ikL) = 1$, de onde $kL = 2 \pi n$, ou seja,

<center>

$k = \frac{2 \pi}{L} n$, com $n=0, \pm 1, \pm 2, \pm 3, ...$ . **(14)**

</center>

Para $L$ muito grande, o intervalo entre dois valores consecutivos de $k$ é muito pequeno (igual a $2\pi/L$) . Podemos, então, substituir uma soma em $k$ por uma integral em $dk$,

<center>

$\sum_{k} f(k) \to \int{\frac{dk}{\frac{2 \pi}{L}} f(k)} = \frac{L}{2 \pi} \int{dk f(k)}.$ **(15)**

</center>

Mais adiante vamos utilizar esse resultado. Apenas precisamos tomar cuidado quando a função $f(k)$ tiver algum tipo de singularidade.

Todos esses resultados podem ser facilmente generalizados para três dimensões. Nesse caso, temos

<center>

$\psi(\overrightarrow{r}) = C exp(i \overrightarrow{k} . \overrightarrow{r})$ **(16)**

</center>

com

<center>

$\overrightarrow{k} = \frac{2 \pi}{L_{1}} m_{1} \overrightarrow{e_{x}} + \frac{2 \pi}{L_{2}} m_{2} \overrightarrow{e_{y}} + \frac{2 \pi}{L_{3}} m_{3} \overrightarrow{e_{z}}$ **(17)**

</center>

onde $m_{1}$, $m_{2}$ e $m_{3}$ são inteiros, $\overrightarrow{e_{x}}$, $\overrightarrow{e_{y}}$ e $\overrightarrow{e_{z}}$ são vetores unitários ao longo das direções cartesianas e a particula está dentro de uma caixa de lados $L_{1}$, $L_{2}$ e $L_{3}$. A energia do orbital será dada por

<center>

$\varepsilon_{\overrightarrow{k}} = \frac{\hbar^{2}k^{2}}{2m}.$ **(18)**

</center>

A transformação de uma soma em uma integral se faz de maneira análoga, mediante a relação

<center>

$\sum_{\overrightarrow{k}} f(\overrightarrow{k}) \to \frac{V}{(2 \pi)^{3}} \int{d^{3} \overrightarrow{k} f(\overrightarrow{k})}$ **(19)**

</center>

Até agora construímos os orbitais com base apenas nas coordenadas de posição. Pode haver, no entanto, outros graus de liberdade: spin, isospin etc. Para levar em conta o spin, podemos caracterizar um orbital por meio do símbolo $j \equiv (\overrightarrow{k}, \sigma)$, que inclui o **vetor de onda** $k$, associado aos **graus de liberdade de translação**, e o **número quântico de spin** $\sigma$. Para partículas livres, na ausência de termos de spin no operador hamiltoniano, o espectro de energia continua sendo dado por

<center>

$\varepsilon_{j} \equiv \varepsilon_{\overrightarrow{k}, \sigma} \equiv \frac{\hbar^{2}k^{2}}{2m}.$ **(20)**

</center>

Na presença de um campo magnético $H$, levando em conta a interação entre o spin e o campo, o espectro de energia de uma partícula livre de spin 1/2 será dado por

<center>

$\varepsilon_{j} \equiv \varepsilon_{\overrightarrow{k}, \sigma} \equiv \frac{\hbar^{2}k^{2}}{2m} - \mu_{B}H\sigma,$ **(21)**

</center>

onde a constante $\mu_{B}$ é o **magneton de Bohr** e a variável de spin $\sigma$ pode assumir os valores $\pm 1$.

A linguagem dos orbitais é muito interessante para tratar situações em que as partículas têm uma estrutura interna complexa, como no caso de um **gás diluído de moléculas diatômicas**. Os modelos clássicos para um gás de moléculas diatômicas (um rotor rígido em três dimensões ou um rotor com eixo vibrante) são incapazes de explicar o comportamento de certas grandezas termodinâmicas, como a dependência do calor específico a volume constante com a temperatura. O hamiltoniano de uma molécula diatômica pode ser escrito na forma de uma soma de vários termos, representando diferentes graus de liberdade, que não se acoplariam em primeira aproximação,

<center>

${\cal{H}_{mol}} = {\cal{H}_{tr}} + {\cal{H}_{el}} + {\cal{H}_{rot}} + {\cal{H}_{vib}} + ...$ . **(22)**

</center>

O primeiro termo representa a translação do centro de massa, caracterizada pelo vetor de onda $\overrightarrow{k}$. O segundo termo se refere aos estados eletrônicos, que podem ser calculados supondo que os núcleos estejam fixos (na chamada aproximação de Born-Oppenheimer). Os estados eletrônicos excitados estão, em geral, muito acima do nivel fundamental e praticamente não contribuem para a energia livre termodinâmica. O termo seguinte, o corresponde às rotações em torno de um eixo normal ao eixo da molécula, com momento de inércia $I$ (as funções de onda são harmônicos esféricos, caracterizados pelo número quântico de momento angular $J$ e pela sua projeção $m_{J}$, com degenerescência $2J + 1$). O termo ${\cal{H}_{vib}}$ representa as vibrações ao longo do eixo da molécula, com uma frequência fundamental $\omega$ associada à forma do mínimo do potencial molecular. Levando em conta apenas esses termos, temos

<center>

$\varepsilon_{j} \equiv \varepsilon_{\overrightarrow{k}, J, n} \equiv \frac{\hbar^{2}k^{2}}{2m} + \frac{\hbar^{2}}{2I} J(J + 1) + \hbar \omega(n + \frac{1}{2}),$ **(23)**

</center>

onde $n$ e $J$ percorrem os números inteiros, a partir de zero. Mais adiante vamos ver algumas consequências desse resultado.

## 3) Formulação do problema estatístico

Devido às propriedades de simetria da função de onda, um estado quântico do gás ideal fica inteiramente caracterizado pelo conjunto de números

<center>

$\{n_{1}, n_{2}, ..., n_{j}, ... \} \equiv \{n_{j}\}, $ **(24)**

</center>

onde $j$ designa o **estado quântico de um orbital** e $n$, é o **número de partículas no orbital** $j$. No caso de férmions, é claro que $n_{j} = 0$ ou $1$, para qualquer $j$. No caso de bósons, no entanto, $n$, pode variar de $0$ a $N$, onde $N$ é o **número total de partículas**. A energia do sistema correspondente ao estado quântico $\{n\}$ é dada por

<center>

$E\{n_{j}\} = \sum_{j} \varepsilon_{j} n_{j}$ **(25)**

</center>

onde $\varepsilon_{j}$ é a **energia do orbital** $j$. O número total de partículas é dado por

<center>

$N = N\{n_{j}\} = \sum_{j} n_{j}.$ **(26)**

</center>

É importante notar que no tratamento estatístico quântico do gás ideal estamos preocupados em saber apenas quantas partículas estão em cada orbital. No caso clássico, representado pelo modelo do gás de Boltzmann, em que as partículas são distinguíveis, precisaríamos saber quais partículas estão em cada orbital.

No **ensemble canônico**, podemos escrever a função de partição do gás ideal quântico na forma

<center>

$Z = Z(T, V, N) = \underset{(\sum_{j} n_{j} = N)}{\sum_{\{n_{j}\}}} exp[- \beta \sum_{j} \varepsilon_{j} n_{j}],$ **(27)**

</center>

onde o problema reside em calcular a soma (que agora não se fatoriza) sobre os números de ocupação com a restrição imposta pelo número fixo de partículas. Essa situação vai ficar consideravelmente mais simples no **ensemble grande canônico** devido à ausência de restrições sobre a conservação do número total de partículas.

No ensemble grande canônico, temos

<center>

$\Xi = \Xi(T, V, \mu) = \sum_{N=0}^{\infty} exp(\beta \mu N) Z(T, V, N) = $ 

$ = \sum_{N=0}^{\infty} exp(\beta \mu N) \underset{(\sum_{j} n_{j} = N)}{\sum_{\{n_{j}\}}} exp[-\beta \varepsilon_{1} n_{1} - \beta \varepsilon_{2} n_{2} - ...] = $

$\sum_{N=0}^{\infty} \underset{(\sum_{j} n_{j} = N)}{\sum_{\{n_{j}\}}} exp[-\beta(\varepsilon_{1} - \mu)n_{1} - \beta(\varepsilon_{2} - \mu)n_{2} - ...].$ **(28)**

</center>

Como a soma é inicialmente feita sobre o conjunto de números de ocupação $n_{1}, n_{2}, ...,$ com a restrição de que $N=n_{1} + n_{2} + ...$ esteja fixo, mas depois há uma outra soma sobre todos os valores de $N$, podemos simplesmente fazer uma soma múltipla sobre todos os números de ocupação, sem qualquer tipo de restrição (o que cor responde a um mero rearranjo dos termos dessa soma!). Então, temos

<center>

$\Xi = \sum_{n_{1}, n_{2}, ...} exp[-\beta(\varepsilon_{1} - \mu)n_{1} - \beta(\varepsilon_{2} - \mu)n_{2} - ...] ,$ **(29)**

</center>

que agora se fatoriza, podendo ser escrita na forma

<center>

$\Xi = \{\sum_{n1} exp[-\beta(\varepsilon_{1} - \mu)n_{1}]\} \{\sum_{n2} exp[-\beta(\varepsilon_{2} - \mu)n_{2}]\} ...,$ **(30)**

</center>

ou seja,

<center>

$\Xi = \Xi(T, V, \mu) = \underset{j}{\Pi} \{\sum_{n} exp[-\beta(\varepsilon_{j} - \mu)]\}.$ **(31)**

</center>

A partir de $E$, usando o formalismo do ensemble grande canónico, podemos obter expressões para os valores esperados da energia e do número de partículas e escrever uma equação para o **grande potencial termodinâmico**. O valor esperado $\langle n_{j} \rangle$ do número de ocupação do orbital $j$ pode ser obtido por meio da relação

<center>

$\langle n_{j} \rangle = -\frac{1}{\beta} \frac{\partial}{\partial \varepsilon_{j}} \ln{\Xi}.$ **(32)**

Cabe agora distinguir entre os dois tipos possíveis de estatísticas.

</center>

**Estatística de Bose-Einstein**

Nesse caso, a soma em $n$ na equação **(31)** varia de $0$ a $\infty$. Então, temos

<center>

$\sum_{n=0}^{\infty} exp[-\beta(\varepsilon_{j} - \mu)n] = \{1 - exp[-\beta(\varepsilon_{j} - \mu)]\}^{-1}.$ **(33)**

</center>

Note que essa soma só pode existir desde que $exp[-\beta (\varepsilon_{j} - \mu)] < 1$ para qualquer $j$. Levando em conta que o menor valor de $\varepsilon_{j}$ é $0$, temos $exp(\beta \mu) < 1$, ou seja, o **potencial quimico** $\mu$ **deve sempre ser negativo**. A situação em que $\mu \to 0-$ é muito interessante, dando origem ao fenômeno da **condensação de Bose-Einstein**. Temos, então,

<center>

$\ln{ \Xi(T, V, \mu)} = -\sum_{j} \ln{\{1 - exp[-\beta(\varepsilon_{j} - \mu)]\}}.$ **(34)**

</center>

A partir da equação **(32)**, obtemos o **número médio de ocupação do orbital** $j$,

<center>

$\langle n_{j} \rangle = \frac{1}{exp[\beta(\varepsilon_{j} - \mu)] - 1}.$ **(35)**

</center>

A condição $exp[-\beta(\varepsilon_{j} - \mu)] < 1$ está, portanto, relacionada à exigência física de que $\langle n_{j} \rangle \geq 0$ para qualquer orbital $j$.

```python
'''
Sabemos que β = 1 / Kb . T

Partindo de <nj> = 1 / exp[β(εj - μ) - 1]

∴ F(εj, μ, T) = 1 / ((e^(εj - μ) / Kb . T) - 1) -> Função de Bose-Einstein

Vemos que a distribuição diverge à medida que se aproxima de μ, ou seja, à medida que
ε − μ se aproxima de 0. Isso significa que à medida que a temperatura se aproxima de zero, 
todas as partículas estarão no estado fundamental.
'''
newparams = {'axes.labelsize': 10, 'axes.linewidth': 1.5, 'savefig.dpi':   
    300,   
    'lines.linewidth': 2, 'figure.figsize': (12, 10),    
    'legend.frameon': True,   
    'legend.handlelength': 0.7}   
plt.rcParams.update(newparams)   

fig, ax = plt.subplots(figsize=(15, 10))
sns.set_theme()

kt = 1.0
legarray = ['FD', 'BE', 'MB']
x = np.linspace(-10, 10)
f = 1.0 / (np.exp(x / kt) + 1)
plt.plot(x, f, 'b')

x = np.linspace(0.4, 10)
f = 1.0 / (np.exp(x / kt) - 1)
plt.plot(x, f, 'g')

x = np.linspace(-1.0, 10)
f = np.exp(-x / kt)
plt.plot(x, f, 'r')

plt.xlabel(r'$(\epsilon-\mu)/K_{B}T$', fontsize=14)
plt.ylabel(r'$(\epsilon, \mu,T)$', fontsize=14)
plt.ylim(0, 2)
plt.xlim(-10, 10)
plt.legend(legarray, facecolor='w', fontsize=14)
plt.title('Plot da Distribuição de Bose-Einstein', fontsize=20)
plt.savefig('Figura1.png')
plt.show()
```


**Estatística de Fermi-Dirac**

Nesse caso, $n$ só pode assumir os valores $0$ ou $1$. Então, temos

<center>

$\sum_{n=0,1} exp[-\beta(\varepsilon_{j} - \mu)n] = 1 + exp[-\beta(\varepsilon_{j} - \mu)],$ **(36)**

</center>

de onde vem que

<center>

$\ln{\Xi(T, V, \mu)} = \sum_{j} \ln{\{1 + exp[-\beta(\varepsilon_{j} - \mu)]\}}.$ **(37)**

</center>

Então,

<center>

$\langle n_{j} \rangle = \frac{1}{exp[\beta(\varepsilon_{j} - \mu)] + 1},$

</center>

podendo-se notar que $0 \geq \langle n_{j} \rangle \geq 1$ em concordância com as exigências do princípio de exclusão de Pauli.

As fórmulas no ensemble grande canónico para os dois tipos de partículas (férmions ou bósons) podem ser resumidas pelas seguintes equações:

<center>

$\ln{\Xi_{FD, BE}} = \pm \sum_{j} \ln{\{1 \pm exp[-\beta(\varepsilon_{j} - \mu)]\}}$ **(38)**

</center>

e

<center>

${\langle n_{j} \rangle}_{FD, BE} = \frac{1}{exp[\beta(\varepsilon_{j} - \mu)] \pm 1},$ **(39)**

</center>

onde $FD$ e $BE$ indicam as estatísticas de Fermi-Dirac e Bose-Einstein, respectivamente, com o sinal $+$ para férmions e o sinal $-$ para bósons. É importante notar que tanto $\Xi$ quanto ${\langle n_{j} \rangle}$ estão expressos em termos das variáveis independentes $T$, $V$ e $\mu$ (que, muitas vezes, nem são as variáveis mais adequadas para o problema fisico sob consideração). Também é importante lembrar que a conexão entre o ensemble grande canônico e a termodinâmica se faz por meio da correspondência


<center>

$\Xi(T, V, \mu) \to exp[-\beta \Phi(T, V, \mu)],$ **(40)**

</center>

no limite termodinâmico $(V \to \infty)$ e que o grande potencial termodinâmico pode ser escrito na forma

<center>

$\Phi(T, V, \mu) = -V_{p}(T, \mu).$ **(41)**

</center>

```python
'''
Sabemos que β = 1 / Kb . T

Partindo de <nj> = 1 / exp[β(εj - μ) + 1]

∴ F(εj, μ, T) = 1 / ((e^(εj - μ) / Kb . T) + 1) -> Função de Fermi

Esta é uma função que fornece uma transição suave de um valor de 1 
quando x << −1 para 0 quando x >> +1. Traçamos fF-D(ε, μ, T) como uma função 
de (ε − μ) / Kb . T; A temperatura determina efetivamente a faixa 
da região de transição. Podemos ilustrar isso introduzindo uma escala de 
energia no gráfico. Vamos usar o potencial químico μ como a escala de energia 1, 
ou seja, plotamos fF-D(ε, μ, T) em função de ε / μ para diferentes valores de T 
correspondentes a kT = 0,2μ, kT = μ e kT = 5µ. Vemos que quando a temperatura vai 
para zero, a função se aproxima de uma função degrau, e quando a temperatura aumenta 
a transição se torna mais ampla.
'''

newparams = {'axes.labelsize': 10, 'axes.linewidth': 1.5, 'savefig.dpi':   
    300,   
    'lines.linewidth': 2, 'figure.figsize': (12, 10),    
    'legend.frameon': True,   
    'legend.handlelength': 0.7}   
plt.rcParams.update(newparams)   

fig, ax = plt.subplots(figsize=(15, 10))
plt.subplot(2, 1, 1)
x = np.linspace(-10, 10)
f = 1.0 / (np.exp(x) + 1)
plt.plot(x, f)
plt.xlabel(r'$(\epsilon-\mu)/K_{B}T$')
plt.ylabel(r'$(\epsilon;\mu,T)$')
plt.xlim(-10, 10)

kbtvalues = [5.0, 1.0, 0.2]
legarray = []
colors = ('blue', 'forestgreen', 'red')
plt.subplot(2, 1, 2)
for i in range(len(kbtvalues)):
    kt = kbtvalues[i]
    x = np.linspace(-20, 20)
    f = 1.0 / (np.exp((x - 1) / kt) + 1)
    plt.plot(x, f, color=f'{colors[i]}')
    legarray.append(r'$K_{B}T$' + f' = {kt:3.1f}')
plt.xlim(-20, 20)
plt.xlabel(r'$\epsilon/\mu$')
plt.ylabel(r'$f(\epsilon, \mu,T)$')
plt.suptitle('Plot da Distribuição de Fermi-Dirac', fontsize=20)
plt.legend(legarray, facecolor='white')
plt.savefig('Figura2.png')
plt.show()
```

## 4) Limite clássico

Observando a expressão $\langle n_{j} \rangle$ para temperaturas baixas ($\beta$ grande) vemos que: (i) no caso de férmions, $\langle n_{j} \rangle$ para $\varepsilon_{j} < \mu$ isto é, para os orbitais com energias mais baixas, e $\langle n_{j} \rangle \approx 0$ para $\varepsilon_{j} > \mu$ (ii) no caso dos bósons, $\langle n_{j} \rangle \approx 0$ para a grande maioria dos orbitais, exceto para as energias mais baixas, quando $\langle n_{j} \rangle \gg 1$.

Na situação clássica não devemos distinguir entre bósons e férmions (isto é, esperamos obter  $\langle n_{j} \rangle \ll 1$ para qualquer $j$). Portanto, o caso clássico deve corresponder a um regime em que

<center>

$exp[\beta(\varepsilon_{j} - \mu)] \gg 1,$ **(42)**

</center>

para qualquer $j$, ou seja,

<center>

$z = exp(\beta \mu) \ll 1.$ **(43)**

</center>

Expandindo as equações **(38)** e **(39)** para $exp(\beta \mu)$ pequeno, temos

<center>

$\ln{\Xi_{FD, BE}} = \sum_{j} exp[-\beta(\varepsilon_{j} - \mu)] + \frac{1}{2} \sum_{j} exp[-2\beta(\varepsilon_{j} - \mu)] + ...$ **(44)**

</center>

e

<center>

${\langle n_{j} \rangle}_{FD, BE} = exp[-\beta (\varepsilon_{j} - \mu)] \{1 + exp[-\beta(\varepsilon_{j} - \mu)] + ...\}.$ **(45)**

</center>

Portanto, o limite clássico fica inteiramente definido pelo termo dominante dessas expressões,

<center>

$\ln{\Xi_{d}} = \sum_{j} exp[-\beta(\varepsilon_{j} - \mu)],$ **(46)**

</center>

e

<center>

${\langle n_{j} \rangle}_{d} = exp[-\beta(\varepsilon_{j} - \mu)].$ **(47)**

</center>

Considerando um espectro de energia dado pela equação **(20)**,

<center>

$\varepsilon_{j} \equiv \varepsilon_{\overrightarrow{k}, \sigma} \equiv \frac{\hbar^{2}k^{2}}{2m}$ **(48)**

</center>

temos

<center>

$\ln{\Xi_{d}} = \sum_{\overrightarrow{k}, \sigma} exp[-\beta(\frac{\hbar^{2}k^{2}}{2m} - \mu)]$ **(49)**

</center>

No limite termodinâmico, a soma nos graus de liberdade de translação pode ser substituída por uma integral. Então, temos

<center>

$\ln{\Xi_{d}} = \gamma \frac{V}{(2\pi)^{3}} \int{d^{3} \overrightarrow{k} exp[-\beta(\frac{\hbar^{2}k^{2}}{2m} - \mu)]} = $ 

$ = \gamma \frac{V}{(2\pi)^{3}} exp(\beta \mu) (\frac{2 \pi m}{\beta \hbar^{2}})^{3/2},$ **(50)** 

</center>

onde $\gamma = 2S + 1$ é a degenerescência dos estados de spin. O grande potencial termodinâmico é dado pela expressão

<center>

$\Phi_{d} = -\gamma V (\frac{2 \pi m}{\beta h^{2}})^{3/2} (k_{B}T)^{5/2} exp(\frac{\mu}{k_{B}T}),$ **(51)**

</center>

justificando a normalização do volume do espaço de fase clássico pela constante de Planck $h$ e a inclusão do fator de contagem correta de Boltzmann (o fator adicional $\gamma$ não tem realmente um análogo clássico).

Escrevendo a equação **(50)** na forma

<center>

$\ln{\Xi_{d}} = \gamma V_{z} (\frac{2 \pi m}{\beta h^{2}})^{3/2},$ **(52)**

</center>

obtemos o número médio de partículas

<center>

$\langle \sum_{j} n_{j} \rangle = z \frac{\partial}{\partial z} \ln{\Xi_{d}} = \gamma V_{z} (\frac{2 \pi m}{\beta h^{2}})^{3/2} \to N$ **(53)**

</center>

que deve ser identificado com o número termodinâmico $N$. Portanto, podemos escrever

<center>

$exp(\beta \mu) = \frac{N}{V} \frac{h^{3}}{\gamma (2 \pi m k_{B} T)^{3/2}}$ **(54)**

</center>

Então, o limite clássico deve corresponder a 

<center>

$\frac{N}{V} \frac{h^{3}}{\gamma (2 \pi m k_{B} T)^{3/2}} \ll 1$ **(55)**

</center>

Para entender o significado dessa última desigualdade, basta observar que

<center>

$(\frac{V}{N})^{1/3} = a$ **(56)**

</center>

representa uma distância intermolecular típica e que

<center>

$\lambda = \frac{h}{\sqrt{2 \pi m k_{B} T}}$ **(57)**

</center>

pode ser associado ao **"comprimento de onda térmico" de de Broglie**. De fato, considerando partículas com uma energia típica $(3/2) k_{B} T$ e uma velocidade típica $v = \sqrt{3 k_{B} T/m}$, temos a **frequência de de Broglie**,

<center>

$v = \frac{3 k_{B} T}{2h},$ **(58)**

</center>

de onde se obtém o comprimento de onda térmico

<center>

$\lambda_{T} = \frac{v}{V} = \frac{2h}{\sqrt{3 k_{B} T m}}$ **(59)**

</center>

que tem a mesma ordem de grandeza do comprimento $\lambda$ dado pela equação **(57)**. Então, no limite clássico devemos ter

<center>

$a = (\frac{V}{N})^{1/3} \gg \lambda_{T} = \frac{2h}{\sqrt{3 k_{B} T m}}$ **(60)**

</center>

de acordo com as expectativas usuais da mecânica quântica. As expressões clássicas devem ser usadas num regime diluído (baixas densidades) e a temperaturas suficientemente altas.

### (A) *Distribuição de Maxwell-Boltzmann*

Utilizando a equação **(46)**, podemos eliminar o potencial químico na equação **(47)** para o valor esperado do número de ocupação $\langle n_{j} \rangle$ do orbital $j$. Vamos escrever a equação **(46)** na forma

<center>

$\ln{\Xi_{d}} = z \sum_{j} exp(-\beta \varepsilon_{j}).$ **(61)**

</center>

Portanto, o número termodinâmico de partículas deve ser dado por

<center>

$N = \ \frac{\partial}{\partial z} \ln{\Xi_{d}}(\beta, z, V) = z \sum_{j} exp(-\beta \varepsilon_{j}).$ **(62)**

</center>

Então, a partir da equação **(47)** temos

<center>

${\langle n_{j} \rangle}_{d} = z exp(-\beta \varepsilon_{j}) = \frac{N exp(-\beta \varepsilon_{j})}{\sum_{j} exp(-\beta \varepsilon_{j})}$ **(63)**

</center>

Portanto,

<center>

$\frac{{\langle n_{j} \rangle}_{d}}{N} = \frac{exp(-\beta \varepsilon_{j})}{\sum_{j} exp(-\beta \varepsilon_{j})},$ **(64)**

</center>

que é a forma discreta da distribuição clássica de Maxwell-Boltzmann. Uma expressão desse mesmo tipo já foi obtida anteriormente, no contexto do modelo de Boltzmann para um gás ideal clássico com energias discretizadas (sem que, no entanto, fosse apresentada qualquer justificativa para a discretização das energias). No limite termodinâmico, temos

<center>

$\frac{{\langle n_{j} \rangle}_{d}}{N} \to p_{0}(v)dv = \frac{{\frac{V}{(2 \pi)^{3}} exp(- {\frac{\beta \hbar^{2} k^{2}}{2m}}) d^{3} \overrightarrow{k}}}{{\frac{V}{(2 \pi)^{3}} \int{exp(- {\frac{\beta \hbar^{2} k^{2}}{2m}}) d^{3} \overrightarrow{k}}}}$

</center>

Fazendo $p = mv = \hbar k$, podemos escrever

<center>

$p_{0}(v) = 4 \pi (\frac{2 \pi k_{B} T}{m})^{-3/2} v^{2} exp(- \frac{mv^{2}}{2 k_{B} T}),$ **(66)**

</center>

como também já foi obtido anteriormente por meio do ensemble canônico clássico.

### (B) *Limite clássico no formalismo de Helmholtz*

A energia livre de Helmholtz dever ser dada pela transformada de Legendre

<center>

$F = \Phi + \mu N,$ **(67)**

</center>

onde o potencial químico é eliminado por meio da equação

<center>

$N = -(\frac{\partial \Phi}{\partial \mu})_{T, V}$ **(68)**

</center>

Utilizando a expressão do grande potencial termodinâmico, dada pela equação **(51)**, temos

<center>

$F = -k_{B} T N \{1 + \ln{\frac{V}{N}} + \frac{3}{2} \ln{T} - \ln{[\frac{1}{\gamma} (\frac{h^{2}}{2 \pi m k_{B}})^{3/2}]}\}.$ **(69)**

</center>

Com $\gamma = 1$ (ou seja, spin nulo), esta é a expressão da energia livre de Helmholtz que se obtém no contexto do ensemble canonico clássico com as correções que foram introduzidas para esta finalidade (o fator de contagem correta $N!$ e a divisão do elemento $dq dp$ do espaço de fase clássico pela constante de Planck $h$).

A entropia, como função da temperatura, do volume e do número de partículas, é dada por 

<center>

$S = -(\frac{\partial F}{\partial T})_{V, N} = N k_{B} \ln{\frac{V}{N}} + \frac{3}{2} k_{B} N \ln{T} + Ns_{0},$ **(70)**

</center>

onde 

<center>

$s_{0} = \frac{5}{2} k_{B} - k_{B} \ln{[\frac{1}{\gamma} (\frac{h^{2}}{2 \pi m k_{B}})^{3/2}]}.$ **(71)**

</center>

Utilizando a equação de estado $pV=Nk_{B}T$, podemos escrever a entropia em termos da temperatura, da pressão e do número de partículas, estabelecendo uma fórmula particularmente útil para comparações com dados experimentais. A expressão **(70)**, com a constante correta dada pela equação **(71)**, é conhecida na literatura como **equação de Sackur-Tetrode**, em consideração aos trabalhos pioneiros da década de 30 que contribuíram para a aceitação definitiva do formalismo canônico da mecânica estatística. Nessa época, já era possível calcular a função de partição canônica de gases diluídos a partir de dados obtidos mediante experiências de espectroscopia. Todas as constantes na expressão da entropia são importantes para as comparações entre a entropia espectroscópica e a entropia termodinâmica, obtida por meio de medidas calorimétricas.

### (C) *Limite clássico da função canônica de partição*

A partir da equação **(61)** podemos escrever

<center>

$\Xi_{d} = exp[z \sum_{j} exp(-\beta \varepsilon_{j})] = \sum_{N = 0}^{\infty} z^{N} Z_{d}(\beta, V, N),$ **(72)**

</center>

onde $Z_{d}(\beta, V, N)$ é a função canônica de partição clássica em termos de $\beta$, do volume e do número de partículas. Utilizando a definição

<center>

$Z_{1} = \sum_{j} exp(-\beta \varepsilon_{j}),$ **(73)**

</center>

temos

<center>

$exp[z Z_{1}] = \sum_{N=0}^{\infty} z^{N} Z_{d}(\beta, V, N).$ **(74)**

</center>

Portanto, o limite clássico da função canônica de partição deve ser dado por 

<center>

$Z_{d} = \frac{1}{N!} Z_{1}^{N} = \frac{1}{N!} [\sum_{j} exp(-\beta \varepsilon_{j})]^{N},$ **(75)**

</center>

que tem exatamente a mesma forma da função canônica de partição do modelo do gás de Boltzmann (incluindo a correção devido ao fator de contagem correta!).

Numa situação em que as partículas tenham uma determinada estrutura interna (como no caso das moléculas poliatômicas), a energia do orbital $j$ pode ser decomposta na forma

<center>

$\varepsilon_{j} = \varepsilon_{\overrightarrow{k}} + \varepsilon_{int} = \frac{\hbar^{2} k^{2}}{2m} + \varepsilon_{int},$ **(76)**

</center>

onde o segundo termo depende dos graus internos de liberdade. Então, temos

<center>

$Z_{1} = \sum_{j} exp(-\beta \varepsilon_{j}) = Z_{int} \sum_{\overrightarrow{k}} exp(-\beta \varepsilon_{\overrightarrow{k}})$ **(77)**

</center>

Portanto,

<center>

$Z_{1} = Z_{int} \frac{V}{(2 \pi)^{3}} \int{d^{3} \overrightarrow{k} exp(- \frac{\beta \hbar^{2} k^{2}}{2m})} = Z_{int} V (\frac{2 \pi m k_{B} T}{h^{2}})^{3/2}.$ **(78)**

</center>

Para partículas sem nenhuma estrutura interna (ou seja, com $Z_{int} = 1$) , recuperamos o resultado para o gás ideal monoatômico clássico,

<center>

$Z_{d} = \frac{1}{N!} V^{N} (\frac{2 \pi m k_{B} T}{h^{2}})^{3N/2},$ **(79)**

</center>

incluindo todas as correções introduzidas anteriormente.


## 5) Gás diluído de moléculas diatômicas

Desprezando as interações entre os diversos graus de liberdade, a função canônica de partição de um gás diluído de moléculas diatômicas pode ser escrita na forma

<center>

$Z = \frac{1}{N!} V^{N} (\frac{2 \pi m k_{B} T}{h^{2}})^{3N/2} Z_{int}^{N},$ **(80)**

</center>

onde $m = m_{1} + m_{2}$ é a **massa total da molécula**. Para a grande maioria das moléculas heteronucleares (isto é, constituídas por dois núcleos distintos), a função de partição interna é dada por

<center>

$Z_{int} = \gamma_{1} \gamma_{2} g_{e} Z_{rot} Z_{vib},$ **(81)**

</center>

onde $\gamma_{i} = (2S_{i} + 1)$, $S_{i}$ é o spin do núcleo $i (i = 1,2)$, $g_{e}$ é a **degenerescência do estado fundamental eletrônico** e as funções canônicas de rotação e de vibração são dadas pelas expressões

<center>

$Z_{rot} = \sum_{J=0}^{\infty} (2J + 1) exp[-\frac{\beta \hbar^{2}}{2I} J(J + 1)]$ **(82)**

</center>

e

<center>

$Z_{vib} = \sum_{J=0}^{\infty} exp[-\beta \hbar \omega (n + \frac{1}{2})] = [2senh(\frac{\beta \hbar \omega}{2})]^{-1},$ **(83)**

</center>

onde o **momento de inércia** $I$ e a **frequência fundamental** $\omega$ estão relacionados com os parâmetros moleculares (podendo, em princípio, ser obtidos através de dados espectroscópicos). Observando estas expressões, podemos definir as temperaturas características de rotação e de vibração,

<center>

$\Theta_{r} = \frac{\hbar^{2}}{2k_{B}I}$ e $\Theta_{v} = \frac{\hbar \omega}{k_{B}}.$ **(84)**

</center>

A equação **(83)**, para o termo vibracional, é idêntica à função canônica de partição do modelo de Einstein, produzindo uma contribuição para o calor específico a volume constante da forma

<center>

$c_{vib} = k_{B} (\frac{\Theta_{v}}{T})^{2} \frac{exp(\Theta_{v}/T)}{[exp(\Theta_{v}/T) - 1]^{2}}.$ **(85)**

</center>

Para baixas temperaturas $(T \ll \Theta_{v})$ o calor específico se anula exponencialmente. Para altas temperaturas $(T \gg \Theta_{r})$ , recuperamos o resultado clássico dado pelo teorema da equipartição, $c_{vib} = k_{B}$ . Na grande maioria das moléculas diatômicas a temperatura característica $\Theta_{v}$ é muito grande, exigindo a utilização dos resultados quânticos para explicar o comportamento experimental.

No limite de baixas temperaturas $(T \ll \Theta_{r})$ , a função de partição rotacional pode ser desenvolvida em série,

<center>

$Z_{rot} = \sum_{J=0}^{\infty} (2J + 1) exp[-\frac{\Theta_{r}}{T} J(J + 1)] = 1 + 3exp(-\frac{2 \Theta_{r}}{T}) + ...$ . **(86)**

</center>

Temos então uma contribuição para a energia livre da forma

<center>

$f_{rot} = - \frac{1}{\beta} \ln{Z_{rot}} = -3 k_{B} T exp(-\frac{2 \Theta_{r}}{T}) + ...$. **(87)**

</center>

Nesse limite de baixas temperaturas, o calor específico rotacional a volume constante também se anula exponencialmente com a temperatura, de acordo com a expressão assintótica

<center>

$c_{rot} = 12 k_{B} (\frac{\Theta_{r}}{T})^{2} exp(- \frac{2 \Theta_{r}}{T}).$ **(88)**

</center>

No limite clássico de altas temperaturas ($T \gg \Theta_{r}$) , o espaçamento entre os níveis rotacionais de energia é muito pequeno. Portanto, substituindo a soma por uma integral na equação **(82)**, recuperamos o resultado clássico, $c_{rot} = k_{B}$ . em concordância com o teorema da equipartição da energia. Nesse regime, no entanto, vamos fazer um cálculo um pouco mais cuidadoso. Dada uma função $\psi(x)$ bem comportada, podemos escrever a expansão de Euler-MacLaurin.

<center>

$\sum_{n=0}^{\infty} \psi_{n}(x) = \int_{0}^{\infty} \psi(x)dx + \frac{1}{2} \psi(0) - \frac{1}{12} \psi'(0) + \frac{1}{720} \psi'''(0) + ...,$ **(89)**

</center>

que estabelece as condições assintóticas para a transformação de uma soma em uma integral. Utilizando essa expansão, podemos escrever a equação **(82)** na forma

<center>

$Z_{rot} = \frac{T}{\Theta_{r}} \{1 + \frac{1}{3} (\frac{\Theta_{r}}{T}) + \frac{1}{15} (\frac{\Theta_{r}}{T})^{2} + ...\}$. **(90)**

</center>

Portanto, a contribuição rotacional para o calor específico a volume constante é dada por

<center>

$c_{rot} = k_{B} \{1 + \frac{1}{45} (\frac{\Theta_{r}}{T})^{2} + ...\}$. **(91)**

</center>

É interessante observar que o calor especifico tende ao valor clássico $k_{B}$ a partir de valores maiores do que $k_{B}$. De fato, no caso geral das moléculas diatômicas, o gráfico experimental da contribuição rotacional para o calor específico em função da temperatura exibe um calombo característico antes de atingir o valor clássico.

### 6.1) Distribuição de Fermi-Dirac, Bose-Einstein e Maxwell-Boltzmann

```python
newparams = {'axes.labelsize': 10, 'axes.linewidth': 1.5, 'savefig.dpi':   
    300,   
    'lines.linewidth': 2, 'figure.figsize': (12, 10),    
    'legend.frameon': True,   
    'legend.handlelength': 0.7}   
plt.rcParams.update(newparams)   
E = np.linspace(-0.5, 0.5, 1001) # Range da energia  
e = 1.6e-19 # Carga elétrica  
k = 1.38e-23 # Constante de Boltzmann (joule per kelvin)  
u = 0 # Considerando o potencial químico da substância é zero

'''
Esta é a equação geral, para distribuição de Maxwell-Boltxmann a = 0,
para Bose-Einstein a = -1 e para Fermi-Dirac a = +1
'''
def Fn(T, a):  
  return 1 / ((np.exp(((E - u) * e) / (k * T))) + a)  

plt.suptitle("Gráfico das seguintes funções em diferentes temperaturas", size=16, color='k', fontstyle ="italic")  

# Distribuição de Maxwell-Boltzmann
plt.subplot(2, 2, 1)
plt.plot(E, Fn(100, 0), label='T = 100 K')  
plt.plot(E, Fn(300, 0), label='T = 300 K')  
plt.plot(E, Fn(500, 0), label='T = 500 K')  
plt.plot(E, Fn(700, 0), label='T = 700 K')  
plt.ylim(0, 3)  
plt.xlabel('Energia (eV)', fontsize=14)  
plt.ylabel('f(E)', fontsize=14)  
plt.legend(loc='best', prop={'size':12}, facecolor='w')  
plt.title("Distribuição de Maxwell-Boltzmann para u=0", fontsize=14)

# Distribuição de Bose-Einstein
plt.subplot(2, 2, 2)  
plt.plot(E, Fn(100, -1), label='T = 100 K')  
plt.plot(E,Fn(300,-1), label='T = 300 K')  
plt.plot(E,Fn(500,-1), label='T = 500 K')  
plt.plot(E,Fn(700,-1), label='T = 700 K')  
plt.xlim(0, 1)  
plt.ylim(0, 2)  
plt.xlabel('Energia (eV)', fontsize=14)  
plt.ylabel('f(E)', fontsize=14)  
plt.legend(loc='best', prop={'size':12}, facecolor='w')        
plt.title("Distribuição de Bose-Einstein para u=0", fontsize=14)

# Distribuição de Fermi-Dirac
plt.subplot(2, 2, 3)  
plt.plot(E,Fn(100, +1), label='T = 100 K')  
plt.plot(E,Fn(300, +1), label='T = 300 K')  
plt.plot(E,Fn(500, +1), label='T = 500 K')  
plt.plot(E,Fn(700, +1), label='T = 700 K')  
plt.legend(loc='best', prop={'size':12}, facecolor='w')  
plt.ylim(0, 1)  
plt.xlabel('Energia (eV)', fontsize=14)  
plt.ylabel('f(E)', fontsize=14)  
plt.title("Distribuição de Fermi-Dirac para u=0", fontsize=14)

plt.subplot(2, 2, 4)  
plt.plot(E,Fn(500, 0), label='Distribuição MB')  
plt.plot(E,Fn(500, -1), label='Distribuição BE')  
plt.plot(E,Fn(500, +1), label='Distribuição FD')  
plt.legend(loc='best', prop={'size':12}, facecolor='w')  

plt.ylim(0, 4) 
plt.xlim(-0.3, 0.5) 
plt.xlabel('Energia (eV)', fontsize=14)  
plt.ylabel('f(E)', fontsize=14)  
plt.title("Temperatura = 500 K", fontsize=14)  
"""###### Ajustando o subplot ######"""  
plt.subplots_adjust(left=0.1,  
        bottom=0.1,   
        right=0.9,   
        top=0.9,   
        wspace=0.4,   
        hspace=0.4)
plt.savefig('Figura3.png')
plt.show()
```
