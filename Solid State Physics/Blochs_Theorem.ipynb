```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

# $\psi_1$

```python
# Define the parameters
A = 1.0
alpha = 0.1
a = 1.0

# Define the function
def psi_1(x):
    return A * np.exp(1j * alpha * x) * (np.sin(2 * np.pi * x / a) + np.cos(2 * np.pi * x / a))

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Calculate corresponding y values for psi_1(x) and psi_1(x + a)
y_values = psi_1(x_values)
y_values_shifted = psi_1(x_values + a)

# Calculate the phase factor
phase_factor = np.exp(1j * alpha * a)

# Check if Bloch's theorem is satisfied
is_bloch_theorem_satisfied = np.allclose(y_values_shifted, phase_factor * y_values)

# Plot the results
f = plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x_values, np.real(y_values), label='Real Part')
plt.plot(x_values, np.real(y_values_shifted), label='Real Part (Shifted)', linestyle='--')
plt.title('Real Part of $\psi_{1}$')
plt.xlabel('x')
plt.ylabel('Real($\psi_{1}$)')
plt.legend(frameon=True, facecolor='white', framealpha=0.7)

plt.subplot(2, 1, 2)
plt.plot(x_values, np.imag(y_values), label='Imaginary Part')
plt.plot(x_values, np.imag(y_values_shifted), label='Imaginary Part (Shifted)', linestyle='--')
plt.title('Imaginary Part of $\psi_{1}$')
plt.xlabel('x')
plt.ylabel('Imag($\psi_{1}$)')
plt.legend(frameon=True, facecolor='white', framealpha=0.7)

plt.tight_layout()
plt.show()

# Check and print if Bloch's theorem is satisfied
print(f"\n\nBloch's theorem satisfied: {is_bloch_theorem_satisfied}")
f.savefig('bloch_1.pdf')
```

# $\psi_3$

```python
# Define the parameters
A = 1.0
a = 1.0

# Define the function
def psi_3(x):
    return A * np.sin(np.pi * x / a)

# Generate x values
x_values = np.linspace(0, 5, 1000)

# Calculate corresponding y values for psi_3(x) and psi_3(x + a)
y_values = psi_3(x_values)
y_values_shifted = psi_3(x_values + a)

# Calculate the phase factor
phase_factor = np.exp(1j * np.pi * a / a)

# Check if Bloch's theorem is satisfied
is_bloch_theorem_satisfied = np.allclose(y_values_shifted, phase_factor * y_values)

# Plot the results
f = plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x_values, y_values, label='Function')
plt.plot(x_values, y_values_shifted, label='Function (Shifted)', linestyle='--')
plt.title('Function $\psi_{3}$')
plt.xlabel('x')
plt.ylabel('$\psi_{3}$')
plt.legend(frameon=True, facecolor='white', framealpha=0.7)

plt.subplot(2, 1, 2)
plt.plot(x_values, np.imag(y_values - phase_factor * y_values_shifted), label='Difference (Imaginary Part)', linestyle='--', color='orange')
plt.title('Imaginary Part Difference')
plt.xlabel('x')
plt.ylabel('Imag($\psi_{3} - e^{i k a} \psi_{3}$)')
plt.legend(frameon=True, facecolor='white', framealpha=0.7)

plt.tight_layout()
plt.show()

# Check and print if Bloch's theorem is satisfied
print(f"\n\nBloch's theorem satisfied: {is_bloch_theorem_satisfied}")
f.savefig('bloch_3.pdf')
```
